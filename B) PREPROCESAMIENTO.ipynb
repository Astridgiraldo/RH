Al observar las bases de datos, en especial las fechas de diligenciamiento de las encuestas de información y satisfacción('datasurvey' y 'surveydata'), se comparan con las fechas de retiro y se encuentra que algunos trabajadores siguieron diligenciando encuestas de información y satisfacción después de la fecha de su retiro. Por tanto, se decide estimar estos casos de inconsistencia y tomar una decisón acerca de ellos

```python
#Filtrar empleados con fechas de encuesta posteriores al retiro
r_2015_copy = r_2015.copy()

empleados_inconsistentes2015 = r_2015_copy[r_2015_copy['retirementdate'] < r_2015_copy['surveydate']]
empleados_inconsistentes2015 = empleados_inconsistentes2015.append(r_2015_copy[r_2015_copy['retirementdate'] < r_2015_copy['datesurvey']])

# Mostrar los empleados con inconsistencias
if not empleados_inconsistentes2015.empty:
    print("Empleados con posibles inconsistencias:")
    print(empleados_inconsistentes2015.head().to_string())
else:
    print("No se encontraron inconsistencias.")
```

##**Tratamiento de nulos final**

Aunque ya se habían tratado los nulos antes de juntar las BD, luego de juntaralas se produjeron nuevas filas con nulos debido a la diferencia de tamaños que tenían originalmente.
Se optó por llenar los nulos de la variable 'attrition' con 'No' ya que los datos faltantes corresponden a personas que no se han retirado de la empresa. El resto de variables ('retirementdate', 'retirementtype', 'resignationreason'), se llenaron con la palabra 'no aplica' debido a que son datos que corresponden a características asociadas al retiro.

```python
# Se llenaron los datos nulos con "No", ya que solo estaban los empleados que si renunciaban
r_2015['attrition']=r_2015['attrition'].fillna('No')
r_2015['retirementdate']=r_2015['retirementdate'].fillna('no aplica')
r_2015['retirementtype']=r_2015['retirementtype'].fillna('no aplica')
r_2015['resignationreason']=r_2015['resignationreason'].fillna('no aplica')
```

```python
# Se llenaron los datos nulos con "No", ya que solo estaban los empleados que si renunciaban
r_2016['attrition']=r_2016['attrition'].fillna('No')
r_2016['retirementdate']=r_2016['retirementdate'].fillna('no aplica')
r_2016['retirementtype']=r_2016['retirementtype'].fillna('no aplica')
r_2016['resignationreason']=r_2016['resignationreason'].fillna('no aplica')
```
